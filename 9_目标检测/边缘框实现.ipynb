定义两种表示之间的转换函数

```python
def box_corner_to_center(boxes):
    """从（左上，右下）到（中间，宽度，高度）""" # 后面这个便于画图
    x1, y1, x2, y2 = boxes[:, 0], boxes[:, 1], boxes[:, 2], boxes[:, 3]
    cx = (x1 + x2) / 2
    cy = (y1 + y2) / 2
    w = x2 - x1
    h = y2 - y1
    boxes = torch.stack((cx, cy, w, h), axis=-1) # 表示按最后一个维度 堆叠
    return boxes
```
